```python
import numpy as np
import networkx as nx
import pandas as pd
from collections import defaultdict
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('all_data_08_12_21.tsv',sep='\t')
```

```python
entityA=df['ENTITYA']
entityB=df['ENTITYB']
two=df[['ENTITYA','ENTITYB']]
key=[]
for i in range(len(two)):
    if two.iloc[i]['ENTITYA'] not in key:
        key.append(two.iloc[i]['ENTITYA'])
for j in range(len(two)):
     if two.iloc[j]['ENTITYB'] not in key:
        key.append(two.iloc[j]['ENTITYB'])

```

```python
#Creating unique ids for each element
temp = defaultdict(lambda: len(temp))
res = [temp[ele] for ele in key]
```

```python
edge_list=[]
for i in range(len(two)):
    b=[key.index(two.iloc[i]['ENTITYA']),key.index(two.iloc[i]['ENTITYB'])]
    edge_list.append(b)
edge_list
```

```python
edge_list[2][1]
```

```python
G=nx.Graph()
G.add_nodes_from([res[0],res[1]])
for i in range(len(edge_list)):
    G.add_edge(edge_list[i][0],edge_list[i][1])
```

```python
nx.draw(G)
plt.show()
```
